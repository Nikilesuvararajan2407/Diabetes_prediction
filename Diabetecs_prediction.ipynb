<h3>importing libararies</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

<h3>data loading</h3>

```python
data=pd.read_csv("health care diabetes.csv")
data.head(4)
```

<h3>exploratory data analysis(EDA)</h3><br>

<h3>feature engineering</h3><br>

```python
x = data.iloc[:, [0, 2, 3, 4, 5, 6, 7, 8]]
y = data.iloc[:, 1]
x.head(4)
```

```python
y.head(4)
```

<h1  style="color:red;">model building</h1>

<h3>train test split</h3><br>

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

<h3>model intialization</h3><br>

```python
model=LinearRegression()
```

<h3>model training</h3><br>

```python
model= model.fit(x,y)
```

```python
y_predict = model.predict(x)
print(y_predict)
```

<h3>hyperparameter tunning</h3><br>

<h1   style="color:red;"> evaluvation and result</h1>

```python
rmse=np.sqrt(mean_squared_error(y,y_predict))#iroot mean squared err claculated ,
r2=model.score(x,y)# finding r2 err
print("rmse:",rmse)
print("root squared error:",r2)
```
